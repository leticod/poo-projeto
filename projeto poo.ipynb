```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
class DataAnalysis:
    def __init__(self, file_path):
        self.data = pd.read_excel(r"/content/Poo project - Copiar.xlsx")
    def quantidade_elementos(self):
        return len (self.data)

    def nome_colunas(self):
        return self.data.columns.tolist()

    def tipo_dados_colunas(self):
        return self.data.dtypes

    def analise_estatistica(self):
        return self.data.describe()

    def ajustar_data(self):
        self.data['Data'] = pd.to_datetime(self.data['Data']).dt.strftime('%Y-%m-%d')

    def avaliacao_grafica(self):
        # Distribuição normal dos índices
        self.data.hist()
        plt.show()

        # Boxplot e identificação de outliers
        self.data.boxplot()
        plt.show()

        # Curva índice vs tempo
        self.data.plot(x='Data', y= 'Total')
        plt.figure(figsize=(14, 7))
        plt.show()
```

```python
# Uso da classe
caminho_arquivo = 'seu_arquivo.excel'
analise = DataAnalysis(caminho_arquivo)
print(analise.quantidade_elementos())
print(analise.nome_colunas())
print(analise.tipo_dados_colunas())
print(analise.analise_estatistica())
analise.ajustar_data()
analise.avaliacao_grafica()
```
